```python
df[df["Event distance/length"].isin(["50km","50mi"])]
```

```python
#now only for the year of 2020
```

```python
df[(df["Event distance/length"].isin(["50km","50mi"]))& (df["Year of event"] == 2020)]
```

```python
df[df['Event name'] == 'Everglades 50 Mile Ultra Run (USA)']['Event name'].str.split('(').str.get(1).str.split(')').str.get(1)
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
#combine\ing all of the filters together
```

```python
df[(df["Event distance/length"].isin(["50km","50mi"]))& (df["Year of event"] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df["Event distance/length"].isin(["50km","50mi"]))& (df["Year of event"] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(15)
```

```python
df2.shape
```

```python
#removing the "USA" out of event name since all of the remaining event entries are USA athletes.
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(14)
```

```python
#next need to clean up the ahtelete's age column
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth'].astype(int)
```

```python
df2.head(10)
```

```python
#removing (h) from performance
```

```python
df2['Athlete performance'].str.split('(').str.get(0)
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(12)
```

```python
#Dropping columns: Athlete club, Athlete year of birth, and Athlete country
```

```python
df2 = df2.drop(['Athlete club', 'Athlete year of birth', 'Athlete country'], axis = 1)
```

```python
df2.head(15)
```

```python
#clean any null values
```

```python
df2.isna().sum()
```

```python
df2[df2['Athlete age category'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.isna().sum()
```

```python
#checking for duplicate values
```

```python
df2[df2.duplicated() == True]
```

```python
df2.reset_index(drop = True)
```

```python
#fixing types
```

```python
df2.loc[:, 'Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2 = df2.rename(columns = {'Year of event': 'year' ,
                            'Event name': 'race_name ', 
                            'Event dates': 'race_day',
                            'Event distance/length': 'race_length',
                            'Event number of finishers': 'num_of_race_finishers',
                            'Athlete gender': 'gender',
                            'Athlete age category': 'age',
                            'Athlete average speed': 'average_speed',
                    
                    
                            
                            
                            
                            
                           
                           
                           
                           
                           
                           })





    
```

```python
df2['average_speed'] = pd.to_numeric(df2['average_speed'], errors='coerce')
```

```python
df2.groupby(['race length', 'gender'])['average_speed'].mean()
```
